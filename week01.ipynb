```python
import pandas as pd
column_name = ['Id', 'Name', 'Age', 'Weight','m0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
df = pd.read_csv('patient_heart_rate.csv', names = column_name)
```

```python
df.head(10)
```

```python
df[['Firtsname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
```

```python
df.head()
```

```python
weight = df['Weight']
for i in range(len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x/2.2)
        weight[i]= y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i]= x
```

```python
df.rename(columns={'Weight': 'Weight_kgs'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.dropna(how= 'all', inplace=True)
df
```

```python
df= df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight_kgs'])
```

```python
#8
df.Firtsname.replace({r'[^\x00-X7F]+':''}, regex=True, inplace=True)
df.Lastname.replace({r'[^\x00-X7F]+':''}, regex=True, inplace=True)
df
```

```python
#9
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
#10


#note
# Làm sạch dữ liệu trong cột Weight_kgs
df['Weight_kgs'] = pd.to_numeric(df['Weight_kgs'], errors='coerce')

# Điền giá trị thiếu bằng giá trị trung vị
df['Weight_kgs'].fillna(df['Weight_kgs'].median(), inplace=True)

```

```python
df
```

```python
#11
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'Firtsname', 'Lastname'], value_name='PulseRate', var_name='gender_time').sort_values(['Id','Age','Weight_kgs','Firtsname','Lastname'])
```

```python
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
df
```

```python
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
```

```python
df = pd.concat([df, df_temp], axis=1)
```

```python
df
```

```python
df = df.drop(['gender_time'], axis= 1)
```

```python
#12
import numpy as np
df = df.replace('-', np.nan).dropna(subset=['PulseRate'])
df.head(10)
```

```python
#13
df['Firtsname'].isnull().sum()
df['Lastname'].isnull().sum()
df['Firtsname'].fillna('Unknown', inplace= True)
df['Lastname'].fillna('Unknown', inplace= True)
```

```python
#14
df = df.reset_index()
```

```python
df
```

```python
#15
df.to_csv('patient_heart_rate_clean.csv')
```
