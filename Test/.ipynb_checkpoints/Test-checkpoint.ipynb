```python
#CAU 3 B) Import thư viện matplotlib
import matplotlib.pyplot as plt

# Tính số lượng học sinh theo giới tính
count = data["Gender"].value_counts()

# Vẽ đồ thị Pie
plt.pie(count, labels=count.index, autopct="%1.1f%%")
plt.title("So sánh mối tương quan giữa Boy và Girl theo số lượng")
plt.show()
```

```python
#CÂU 5

count_df = data.groupby('REGION')['Gender'].count()
count_df
```

```python
# Import thư viện matplotlib
import matplotlib.pyplot as plt


# Thống kê bảng theo Vùng và Số lượng học sinh
summary = data.groupby("REGION")["Gender"].count()

# Vẽ biểu đồ bar
plt.bar(summary.index, summary)
plt.xlabel("Vùng")
plt.ylabel("Số lượng học sinh")
plt.title("So sánh số lượng học sinh theo Vùng")
plt.show()
```
